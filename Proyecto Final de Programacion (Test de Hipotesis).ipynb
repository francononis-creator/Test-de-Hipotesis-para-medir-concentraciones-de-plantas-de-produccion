```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_1samp, ttest_ind
```

```python
muestra_1 = np.array([9.88, 10.12, 9.95, 10.22, 9.90, 9.85, 10.05, 10.31, 9.99, 10.14,
    9.92, 10.08, 10.27, 9.86, 10.10, 10.03, 9.97, 10.19, 9.89, 10.24,
    10.06, 9.91, 10.15, 9.93, 10.11, 10.04, 9.87, 10.20])

muestra_2 = np.array([9.72, 9.80, 9.88, 9.90, 9.85, 9.78, 9.83, 9.91, 9.76, 9.82,
    9.87, 9.79, 9.89, 9.84, 9.81, 9.77, 9.92, 9.86, 9.75, 9.93,
    9.88, 9.82, 9.79, 9.85, 9.83, 9.90, 9.78, 9.74])
```

```python
#Descriptiva muestra 1
n1 = len(muestra_1)
promedio_1 = muestra_1.mean()
desvio_estandar_1 = np.std(muestra_1)
minimo_1 = min(muestra_1)
maximo_1 = max(muestra_1)
```

```python
print("Estadistica Descriptiva de la muestra de la planta 1: ")
print(f"Posee una cantidad de {n1} registros de datos.")
print(f"Su promedio es: {promedio_1}")
print(f"Su desvio estandar es: {desvio_estandar_1}")
print(f"Su valor menor es {minimo_1} y el mayor {maximo_1}.")
```

```python
#Descriptiva muestra 2
n2 = len(muestra_2)
promedio_2 = muestra_2.mean()
desvio_estandar_2 = np.std(muestra_2)
minimo_2 = min(muestra_2)
maximo_2 = max(muestra_2)
```

```python
print("Estadistica Descriptiva de la muestra de la planta 2: ")
print(f"Posee una cantidad de {n2} registros de datos.")
print(f"Su promedio es: {promedio_2}")
print(f"Su desvio estandar es: {desvio_estandar_2}")
print(f"Su valor menor es {minimo_2} y el mayor {maximo_2}.")
```

Primeras conclusiones comparando solo datos de estadistica descriptiva:
1) Ambos conjuntos de datos poseen 28 registros.
2) El promedio de la muestra uno es superior a la segunda
3) El desvio estandar de la primera muestra es bastante mayor que la segunda, por lo que podemos concluir que sus registros estan más dispersos respecto de la media.
4) Evaluando sus maximos y minimos podemos ver que efectivamente son bastante más dispersos y con valores más elevados.

# Prueba de Hipotesis de la Planta 1

```python
mu = 10.0
```

```python
t1, p1 = ttest_1samp(muestra_1, mu)
```

```python
print(f"Planta 1: t = {t1:.4f}, p = {p1:.4f}")
```

# Conclusion de la primera planta: 
En la primera planta el p-valor es mayor que el nivel de significancia (α = 0.05). 
Por ende no se encontraron evidencias suficientes para rechazar la hipótesis nula, por lo que el promedio puede considerarse compatible con el valor normativo de 10.

# Prueba de Hipotesis de la Planta 2

```python
t2, p2 = ttest_1samp(muestra_2, mu)
```

```python
print(f"Planta 2: t = {t2:.4f}, p = {p2:.4f}")
```

# Conclusion de la segunda planta: 
En la segunda planta el p-valor es menor que el nivel de significancia (α = 0.05).
Por ende si se encontraron evidencias suficientes para rechazar la hipótesis nula, indicando que el promedio difiere significativamente de 10 y, por lo tanto, no cumple con la normativa.
